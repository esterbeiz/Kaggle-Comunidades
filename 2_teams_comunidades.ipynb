## Columnas del dataframe:

   -change: Cambio en la posición del equipo.
   
   -competitionId: Id de la competición.
   
   -entries: Número de entries por equipo en cada una de las competiciones.
   
   -lastSubmission: Fecha de cierre de la competición
   
   -medal rank: Tipo de medalla obtenida por el equipo (gold, silver, bronze). NaN, equipos que no obtienen medalla.
   
   -score: Puntuación obtenida por el equipo en la competición. Hace referncia a la puntuación final, una vez que conocemos la variación de su posición en la competición.
   
   -sourceKernelName: Hace referencia al nombre del cada uno de los kernel (https://www.kaggle.com/chenglongchen/prudential-seed-23/data). Casi el 97% de los datos son nulos, por lo que no parece un dato muy significativo a la hora del análisis.
   
   -sourceKernelUrl: Identificadción la URL del kernel. Al igual que en 'sourceKernelName' casi el 97% de los datos son nulos.
   
   -teamId: Id identificativo del equipo.
   
   -teamMembers: Miembros que forman cada uno de los equipos.
   
   -teamName: Nombre del equipo

```python
df['lastSubmission'] = pd.to_datetime(df['lastSubmission'])
```

```python
# Create features for year, month, day, hour, and minute 
df['year'] = df['lastSubmission'].dt.year 
df['month'] = df['lastSubmission'].dt.month 
df['day'] = df['lastSubmission'].dt.day 
df['hour'] = df['lastSubmission'].dt.hour 
df['minute'] = df['lastSubmission'].dt.minute
```

```python
df.head()
```

```python
#Ordenamos el número de equipos participantes por año.
df_team = df[['year','teamName']].groupby('year').count()
df_team.reset_index(level=0, inplace=True)
df_team
```

```python
#Con ello realizamos un gráfico que muestra la evolución de equipos participantes por año.
plt.figure(figsize=(30,15))
plt.plot(df_team['year'],df_team['teamName'])
plt.title('Equipos participantes por año',fontsize=30)
plt.xlabel('Años',fontsize = 20)
plt.ylabel('Equipos',fontsize = 20)
```
