```python
df['lastSubmission'] = pd.to_datetime(df['lastSubmission'])
```

```python
# Create features for year, month, day, hour, and minute 
df['year'] = df['lastSubmission'].dt.year 
df['month'] = df['lastSubmission'].dt.month 
df['day'] = df['lastSubmission'].dt.day 
df['hour'] = df['lastSubmission'].dt.hour 
df['minute'] = df['lastSubmission'].dt.minute
```

```python
df.head()
```

```python
#Ordenamos el nÃºmero de equipos participantes por aÃ±o.
df_team = df[['year','teamName']].groupby('year').count()
df_team.reset_index(level=0, inplace=True)
df_team
```

```python
#Con ello realizamos un grÃ¡fico que muestra la evoluciÃ³n de equipos participantes por aÃ±o.
plt.figure(figsize=(30,15))
plt.plot(df_team['year'],df_team['teamName'])
plt.title('Equipos participantes por aÃ±o',fontsize=30)
plt.xlabel('AÃ±os',fontsize = 20)
plt.ylabel('Equipos',fontsize = 20)
```
