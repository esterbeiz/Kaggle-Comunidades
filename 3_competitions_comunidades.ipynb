### En el tiempo

Separamos la columna 'dealine' para obetner años, meses, días, horas y minutos en columnas diferentes.


```python
df2['deadline'] = pd.to_datetime(df2['deadline'])
```

```python
# Create features for year, month, day, hour, and minute 
df2['year'] = df2['deadline'].dt.year 
df2['month'] = df2['deadline'].dt.month 
df2['day'] = df2['deadline'].dt.day 
df2['hour'] = df2['deadline'].dt.hour 
df2['minute'] = df2['deadline'].dt.minute 
```

```python
df2.head()
```

¿Cómo ha evolucionado el número de competiciones a lo largo de los años?

```python
df2.year.unique()
```

Tenemos datos desde el 2010 hasta el 2018.

```python
#no vale para nada
count = df2['year'].isna().count()
print(count)
```

```python
df_compperyear = df2[['year','competitionId']].groupby('year').count()
df_compperyear.reset_index(level=0, inplace=True)
df_compperyear
```

```python
plt.figure(figsize=(30,15))
plt.plot(df_compperyear['year'],df_compperyear['competitionId'])
plt.title('Competiciones totales por año',fontsize=30)
plt.xlabel('Años',fontsize = 20)
plt.ylabel('Competiciones',fontsize = 20)
```
